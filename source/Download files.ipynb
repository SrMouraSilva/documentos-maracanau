```python
import re

def denied(request):
    return 'permission-denied.txt' in request.headers['Content-disposition']

def filename(request):
    content_disposition = request.headers['Content-disposition']
    name_with_marks = re.search(r'"(.*)"', content_disposition).group(0)
    return name_with_marks[1:-1]


#start_number = 1
#end_number = 20000

start_number = 571
end_number = 30000
#end_number = 30000


for i in range(start_number, end_number):
    try:
        r = requests.get(f'https://camaramaracanau.ce.gov.br/download/?wpdmdl={i}')
        print(i, end=' ')
        if denied(r):
            continue
        name = filename(r)
        print(i, name)
        open(f'../documentos-originais/{i}-{name}', 'wb').write(r.content)
    except Exception as e:
        print('\n============================')
        print(' ERRO', i, e)
        print('============================')
```
